### 7. Age

```python
def calculate_media_description(media):
    if media['Age'] < 19:
        if media['POT'] > 80:
            return 'Wonderkid'
        return 'Very Young'
    elif 20 <= media['Age'] <= 24:
        return 'Youngster'
    elif 25 <= media['Age'] <= 30:
        if media['↓OVA'] > 85:
            return 'World Class'
        return 'Prime'
    else:
        return 'Old'

# Apply the function row-wise and insert the new column beside 'Age'
fifa.insert(
    fifa.columns.get_loc('Age') + 1, 
    'Media Description',            
    fifa.apply(calculate_media_description, axis=1) 
)

# Display the updated DataFrame
print(fifa[['Age', 'Media Description', 'POT', '↓OVA']])
```

### 13. Height

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches*2.54)

# Apply fn to height column

fifa['Height'] = fifa ['Height'].apply(convert_height)
fifa['Height'].unique()
```

```python
# Rename Column

fifa = fifa.rename(columns = {'Height': "Height (cm)"})
fifa.sample(1)
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip("lbs"))
        return round(pounds/2.205)

# Apply fn to weight column

fifa['Weight'] = fifa ['Weight'].apply(convert_weight)
fifa['Weight'].unique()
```

```python
def convert_value(value):
    value = value.replace('€', '').strip()
    if 'M' in value:
        return round(float(value.replace('M', '')) * 1000000) 
    elif 'K' in value:
        return round(float(value.replace('K', '')) * 1000)
    elif value == '0':
        return 0
    return int(value) 

# Apply the function to the 'Approx Values' column
fifa['Approx Values'] = fifa['Approx Values'].apply(convert_value)
```
